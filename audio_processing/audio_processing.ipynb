```python

import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
from scipy.io.wavfile import write, read
```

```python
# sampling frequency
fs = 44100 

# duration of recording
seconds = 3 

myrecording = sd.rec(int(seconds * fs), samplerate=fs, channels=2)

# wait until recording is finished
sd.wait()  

# save as WAV file 
write('output3.wav', fs, myrecording) 
```

```python
# load audio file
audio_path = "output1.wav"
audio = read(audio_path)
fs = audio[0]
t = np.linspace(0, len(audio[1]) / fs, num=len(audio[1]))

plt.plot(t, audio[1])
plt.show()
```

```python
# playing around with generated sine wave
period = 100
amplitude = 100
data = amplitude * np.sin(period * t)
plt.plot(t, data)
plt.show()

sd.play(data, fs)

# wait until file is done playing
sd.wait()  
```
